```python
column_names = [f'Column_{i}' for i in range(0, 65)]
```

```python
df_one.columns = column_names
df_two.columns = column_names
df_three.columns = column_names
df_four.columns = column_names
df_five.columns = column_names
df_six.columns = column_names
df_seven.columns = column_names
df_eight.columns = column_names
df_nine.columns = column_names
```

```python
master_df=pd.concat([df_one,df_two,df_three,df_four,df_five,df_six,df_seven,df_eight,df_nine]).reset_index(drop=True)
```

```python
master_df.shape
```

```python
master_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
y = master_df["Column_64"]
```

```python
y.shape
```

```python
master_df.shape
```

```python
master_df=master_df.drop("Column_64", axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(master_df, y, test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model = SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(predictions)
```

```python
percentage = model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
res = confusion_matrix(y_test, predictions)
print("Confusion Matrix")
print(res)
print(f"Test Set: {len(X_test)}")
print(f"Accuracy = {percentage*100} %")
```
